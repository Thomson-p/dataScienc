```python
import joblib
import pandas as pd


loaded_model = joblib.load(r'/content/accidents_severity.pkl')# Load the machine learning model

newData = pd.DataFrame({'Mist/fog - Total Acc. - 2014': [15, 78, 20, 69],
                        'Cloudy - Total Acc. - 2014': [120, 74, 98, 29],
                        'Hail/Sleet - Total Accidents - 2016': [201, 259, 121, 79],
                        'Dust Storm - Total Accidents - 2016': [56, 259, 138, 199]})# Creating a DataFrame with hypothetical data

# Make predictions using the loaded model
preditions = loaded_model.predict(newData)

# Define thresholds for mapping
low_threshold = 21
high_threshold = 70

#print the predictions with values
results = []
for predition in preditions:
    if predition <= low_threshold:
        mapped_prediction = 'low'
    elif predition <= high_threshold:
        mapped_prediction = 'high'
    else:
        mapped_prediction = 'very high'
    results.append((predition, mapped_prediction))

for predition, mapped_prediction in results:
    print(f'Predicted Outcome: {predition}, Occurrence : {mapped_prediction}')
```
